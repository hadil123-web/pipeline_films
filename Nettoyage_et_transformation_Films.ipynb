***Nettoyage et transformation***




```python
import pandas as pd
import re
import string
```

```python
file_path = "/content/sample_data/avis_films_bruts (1).csv"
df = pd.read_csv(file_path)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)  # Retirer balises HTML
    text = re.sub(r"http\S+|www\S+", " ", text)  # Retirer liens
    text = text.translate(str.maketrans("", "", string.punctuation))  # Retirer ponctuation
    text = re.sub(r"\s+", " ", text).strip()  # Nettoyer espaces
    return text

df["clean_review"] = df["review"].apply(clean_text)
print(df.head())
```

```python
before = len(df)
df = df.drop_duplicates(subset=["film_title", "clean_review"])# supprimer les doublons
after = len(df)
print(f"Nombre de doublons supprimés: {before - after}")
```

```python
# sentiment analysis selon la note
def assign_sentiment(note):
    if note < 3:
        return "Négatif"
    elif note == 3:
        return "Neutre"
    else:
        return "Positif"

df["sentiment"] = df["user_note"].apply(assign_sentiment)
```
